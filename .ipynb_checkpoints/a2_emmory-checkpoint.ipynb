## Part B

```python
import math
import torch
import gpytorch
import gc
import numpy as np
from scipy import optimize
from matplotlib import pyplot as plt
from IPython.display import display, clear_output

%matplotlib inline
%load_ext autoreload
%autoreload 2

torch.set_default_tensor_type('torch.cuda.FloatTensor')
torch.manual_seed(40)
np.random.seed(40)
```

```python
def objective(x):
    return torch.cos(x)**2

domain = torch.linspace(0,5,200)
train_x = torch.tensor(sorted(np.random.choice(np.linspace(0,5,200),10)))
train_y = objective(train_x)
test_x = domain
```

```python
class ExactGPModel(gpytorch.models.ExactGP):
    def __init__(self, train_x, train_y, likelihood, nu, l):
        super(ExactGPModel, self).__init__(train_x, train_y, likelihood)
        self.mean_module = gpytorch.means.ConstantMean()
        self.covar_module = gpytorch.kernels.MaternKernel(nu=nu)
        self.covar_module.lengthscale = l

    def forward(self, x):
        mean_x = self.mean_module(x)
        covar_x = self.covar_module(x)
        return gpytorch.distributions.MultivariateNormal(mean_x, covar_x)
```

```python
def train_and_plot(nu, l):
    torch.no_grad()
    torch.cuda.empty_cache()
    gc.collect()
    
    # initialize likelihood and model
    likelihood = gpytorch.likelihoods.GaussianLikelihood().cuda()
    model = ExactGPModel(train_x, train_y, likelihood, nu, l).cuda()

    model.train()
    likelihood.train()

    # Use the adam optimizer
    optimizer = torch.optim.Adam(model.parameters(), lr=0.1)

    # "Loss" for GPs - the marginal log likelihood
    mll = gpytorch.mlls.ExactMarginalLogLikelihood(likelihood, model)

    for i in range(100):
        # Zero gradients from previous iteration
        optimizer.zero_grad()
        # Output from model
        output = model(train_x)
        # Calc loss and backprop gradients
        loss = -mll(output, train_y)
        loss.backward()
        optimizer.step()

    def plot_observation(prior=False):
        with torch.no_grad(), gpytorch.settings.prior_mode(prior):
            # Make predictions
            observed_pred = likelihood(model(test_x.double()))

            # Initialize plot
            f, ax = plt.subplots(1, 1, figsize=(8, 6))

            # Get upper and lower confidence bounds
            lower, upper = observed_pred.confidence_region()
            # Plot training data as black stars
            ax.plot(train_x.cpu().numpy(), train_y.cpu().numpy(), 'k*')
            # Plot predictive means as blue line
            ax.plot(test_x.cpu().numpy(), observed_pred.mean.cpu().numpy(), 'b')
            # Shade between the lower and upper confidence bounds
            ax.fill_between(test_x.cpu().numpy(), lower.cpu().numpy(), upper.cpu().numpy(), alpha=0.5)
            ax.set_ylim([-3, 3])
            ax.legend(['Observed Data', 'Mean', 'Confidence'])
        
    model.eval()
    likelihood.eval()
    plot_observation(True)
    plot_observation(False)

```

```python
train_and_plot(3/2,1)
```

```python
train_and_plot(5/2,1)
```

```python
train_and_plot(3/2,1/10)
```

```python
train_and_plot(5/2, 1/10)
```
